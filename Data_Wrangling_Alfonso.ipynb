Vamos a eliminar las columnas de telefono y domicilio y vemos que ya no estan

```python
df1=df.drop(labels=['DOMICILIO','TELEFONO'], axis=1)
```

```python
print(df1.shape)
```

Para la columna de TIPO DOC, voy a reemplazar con la moda, o sea el valor mas utilizado, que es 04 y verificar al final que este ok

```python
df1['TIPO DOC'].isnull().sum()
```

```python
modatipodoc = df1['TIPO DOC'].mode()[0]
modatipodoc
```

```python
df1['TIPO DOC'] = df['TIPO DOC'].fillna(modatipodoc)
```

```python
df1['TIPO DOC'].isnull().sum()
```

Para la columna NRO DOC voy a utilizar un calculo de mediana entre aquellos nros que sean del mismo tipo de doc que estoy colocando (o sea 4.0)

```python
df1['NRO DOC'].isnull().sum()
```

```python
df1['NRO DOC'] = df1['NRO DOC'].fillna(df1['NRO DOC'].median())
```
